```python
x = [-2,4,6]
```

```python
y = [3,6,9]
```

```python
fig = plt.figure(figsize=(15,5))

plt.plot(x, y)
plt.xlim(0, 8)
plt.ylim(1, 10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('line plot')
plt.suptitle('Sales Comaprison',size=30,y=1.5)
```

```python
fig = plt.figure(figsize=(15,3))

plt.plot(x, y)
plt.xlim(0, 10)
plt.ylim(-3, 8)
plt.xlabel('X Axis')
plt.ylabel('Y axis')
plt.title('Line Plot',size=10)
plt.suptitle('Sales Comaprison',size=30,y=1.5)
```
