```python
from nltk.tokenize import word_tokenize
```

```python
sent=[]
for i in data["Text"]:
    sent.append(''.join(i))
    
```

```python
sent
```

```python
tokenized_words = word_tokenize(str(sent))
```

```python
tokenized_words
```

```python
words=[]
waste=[]
for j in len(str(tokenized_words)):
    for i in tokenized_words:
        if (i=='@'):
            waste.append(i)
            waste.append(i)[j+1]
        else:
            words.append(i)
    
```
